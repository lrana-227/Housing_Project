```python
import pandas as pd
import os
import csv

pd.options.display.float_format = '{:.2f}'.format
```

```python
merged_df = pd.read_csv('Housing_data_clean_merged.csv')
merged_df.head()
```

```python
years2 = ['2010','2011','2012','2013','2014','2015','2016','2017']
```

```python
state_df = pd.DataFrame(merged_df.groupby("State")[years2].mean())
state_df
```

```python
state_df["2010-2017 Average"] = state_df.mean(axis=1)
state_df = state_df.reset_index()
state_df
```
