```python
import pandas as pd
import os
import csv

pd.options.display.float_format = '{:.2f}'.format
```

```python
inventory_file_name = os.path.join('Raw Data','Property tax data','RDC_Inventory_Hotness_Metrics_County_History.csv')
inventory_file = pd.read_csv(inventory_file_name)

inventory_file_df = pd.DataFrame(inventory_file)
inventory_file_df
```

```python
inventory_df = inventory_file_df[['month_date_yyyymm','hotness_rank','county_name','supply_score','demand_score','median_days_on_market','median_listing_price']]
inventory1__clean_df = inventory_df.dropna(how ='any')
inventory__clean_df= inventory1__clean_df.reset_index(drop = True)
inventory__clean_df
```

```python
inventory_nj_df = inventory__clean_df.loc[inventory__clean_df['county_name'].str.contains('nj')].reset_index(drop = True)

inventory_nj_df['month_date_yyyymm'] = inventory_nj_df['month_date_yyyymm'].astype(int)/100
inventory_nj_df['month_date_yyyymm'] = inventory_nj_df['month_date_yyyymm'].astype(str)
inventory_nj_df[['Year','Month']] = inventory_nj_df['month_date_yyyymm'].str.split('.',expand=True)

inventory_nj_df=inventory_nj_df.astype({'hotness_rank':int,'median_days_on_market':int})
inventory_nj_df[['county_name','State']] = inventory_nj_df['county_name'].str.split(',',expand=True)

inventory_nj_df['county_name']=inventory_nj_df['county_name'].str.capitalize() 
del inventory_nj_df['month_date_yyyymm']
del inventory_nj_df['State']

#inventory_nj_df.drop(['month_date_yyyymm', 'State','Month'], axis = 1,inplace=True) 

inventory_nj_df = inventory_nj_df.rename(columns = {"hotness_rank":"Hotness Rank",
                                                   "county_name":"County",
                                                   "supply_score":"Supply Score",
                                                   "demand_score":"Demand Score",
                                                   "median_days_on_market":"Days on market(Median)",
                                                   "median_listing_price":"Listing Price(Median)"})


inventory_nj_df = inventory_nj_df.sort_values(['Hotness Rank'])
inventory_nj_clean_df = inventory_nj_df.reset_index(drop=True)
inventory_nj_clean_df.head()
```

```python
inventory_output_file_name = os.path.join('Clean Data','Inventory_Hotness_Metrics_NJ.csv')
inventory_nj_df.to_csv(inventory_output_file_name,index = False)
```
